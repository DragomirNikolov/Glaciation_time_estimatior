```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the parameters for the function
t_star = 0.5
epsilon = 0.2

# Define the IF function
def IF(t, t_star, epsilon):
    if t < t_star - epsilon:
        return 0
    elif t_star - epsilon <= t <= t_star + epsilon:
        return (t - t_star + epsilon) / (2 * epsilon)
    else:
        return 1

# Create data for plotting
t_values = np.linspace(0, 1, 500)
if_values = [IF(t, t_star, epsilon) for t in t_values]

# Set the Seaborn theme
sns.set_theme(style="white")

# Create the plot
plt.figure(figsize=(8, 5))


# Set axis limits and labels
plt.ylim(-0.015, 1.05)
plt.xlim(0, 1)
plt.xticks([])  # Remove numbers on the x-axis
plt.yticks([0, 0.5, 1], ["0", "", "1"])  # Custom ticks for the y-axis
plt.title("Limiting Glaciation", fontsize=14)

# Add arrowed axes
# plt.axhline(y=-0.01, color='black', linewidth=1, linestyle='-')
# plt.axvline(x=0, color='black', linewidth=1, linestyle='-')
plt.annotate('', xy=(1, -0.01), xytext=(0, -0.01), arrowprops=dict(facecolor='black',width = 2))  # X-axis arrow
plt.annotate('', xy=(0, 1.05), xytext=(0, -0.-0.01), arrowprops=dict(facecolor='black' ,width = 2))  # Y-axis arrow
plt.plot(t_values, if_values, label='IF(t)', color='b')
# Remove unnecessary borders
sns.despine(left=True, bottom=True)

# Show the plot
plt.show()
```
